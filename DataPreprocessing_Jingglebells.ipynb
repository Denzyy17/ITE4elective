```python
import re

def standardize_label(text):
    """
    Standardizes a given text by converting to lowercase,
    removing leading/trailing whitespace, and replacing multiple spaces with a single space.
    """
    if isinstance(text, str):
        text = text.lower() # Convert to lowercase
        text = text.strip() # Remove leading/trailing whitespace
        text = re.sub(r'\s+', ' ', text) # Replace multiple spaces with a single space
    else:
        text = str(text) # Ensure non-string inputs are converted to string
        text = text.lower() # Convert to lowercase
        text = text.strip() # Remove leading/trailing whitespace
        text = re.sub(r'\s+', ' ', text) # Replace multiple spaces with a single space
    return text

print("The 'standardize_label' function has been defined.")
```

```python
df['Ground Label_standardized'] = df['Ground Label'].apply(standardize_label)
print("Original 'Ground Label' column:")
print(df['Ground Label'].value_counts())
print("\nStandardized 'Ground Label_standardized' column:")
print(df['Ground Label_standardized'].value_counts())
```

```python
import re
import emoji

def remove_html_tags(text):
    """
    Removes HTML tags from a string.
    Handles non-string inputs by converting them to string.
    """
    if not isinstance(text, str):
        text = str(text)
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)

def remove_emojis(text):
    """
    Removes emojis from a string.
    Handles non-string inputs by converting them to string.
    """
    if not isinstance(text, str):
        text = str(text)
    return emoji.replace_emoji(text, replace='')

print("The 'remove_html_tags' and 'remove_emojis' functions have been defined.")
```

```python
print("Comparison of Original and Standardized 'Ground Label' columns (first 5 rows):")
display(df[['Ground Label', 'Ground Label_standardized']].head())

print("\nValue counts for original 'Ground Label' column:")
print(df['Ground Label'].value_counts())

print("\nValue counts for standardized 'Ground Label_standardized' column:")
print(df['Ground Label_standardized'].value_counts())
```

```python
import string

def tokenize_text(text):
    """
    Tokenizes text by converting to lowercase, removing punctuation,
    and splitting into individual words.
    Handles non-string inputs by converting them to string.
    """
    if not isinstance(text, str):
        text = str(text)

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Split into tokens
    tokens = text.split()

    return tokens

print("The 'tokenize_text' function has been defined.")
```

```python
df_cleaned['User Review Title_tokenized'] = df_cleaned['User Review Title_cleaned'].apply(tokenize_text)
df_cleaned['User Review Body_tokenized'] = df_cleaned['User Review Body_cleaned'].apply(tokenize_text)

print("Tokenization applied to 'User Review Title_cleaned' and 'User Review Body_cleaned'.")
```

```python
from langdetect import detect, DetectorFactory

# Set seed for reproducibility, as langdetect is non-deterministic
DetectorFactory.seed = 0

def get_language(text):
    try:
        return detect(str(text))
    except: # Catch any exception, especially those for very short or non-text content
        return 'unknown'

df_cleaned['Detected Language'] = df_cleaned['User Review Body_cleaned'].apply(get_language)
print("Language detection applied to 'User Review Body_cleaned'.")
print("First 5 detected languages:")
print(df_cleaned[['User Review Body_cleaned', 'Detected Language']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine the cleaned title and body for vectorization
df_cleaned['combined_text'] = df_cleaned['User Review Title_cleaned'].fillna('') + ' ' + \
                            df_cleaned['User Review Body_cleaned'].fillna('')

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limiting features to 5000 for practicality

# Fit and transform the combined text data
X_tfidf = tfidf_vectorizer.fit_transform(df_cleaned['combined_text'])

print("TF-IDF vectorization applied to combined text data.")
print(f"Shape of TF-IDF features: {X_tfidf.shape}")
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Ensure 'Ground Label_standardized' exists in df_cleaned
# This column was added to 'df' after 'df_cleaned' was created as a copy.
df_cleaned['Ground Label_standardized'] = df_cleaned['Ground Label'].apply(standardize_label)

# Define features (X) and target (y)
X = X_tfidf
y = df_cleaned['Ground Label_standardized'] # Using df_cleaned to align with combined_text

# Display original class distribution
print("Original class distribution:", Counter(y))

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE to the dataset
X_resampled, y_resampled = smote.fit_resample(X, y)

# Display new class distribution after SMOTE
print("\nNew class distribution after SMOTE:", Counter(y_resampled))
print(f"Shape of X_resampled after SMOTE: {X_resampled.shape}")
```

```python
numerical_features = df_cleaned[['Business Average Rating', 'User Review Rating', 'User Review Count']]

# Apply SMOTE to the numerical features using the same 'smote' object and original 'y' labels
# The 'y' in the previous step was based on df_cleaned, which is consistent here.
# We need to use the original y for fitting SMOTE.
numerical_features_resampled, _ = smote.fit_resample(numerical_features, y)

print("Original shape of numerical features:", numerical_features.shape)
print("Shape of resampled numerical features after SMOTE:", numerical_features_resampled.shape)
print("First 5 rows of resampled numerical features:\n", numerical_features_resampled.head())
```

```python
df_cleaned['title_char_count'] = df_cleaned['User Review Title_cleaned'].apply(lambda x: len(str(x)))
df_cleaned['title_word_count'] = df_cleaned['User Review Title_tokenized'].apply(lambda x: len(x) if isinstance(x, list) else 0)
df_cleaned['body_char_count'] = df_cleaned['User Review Body_cleaned'].apply(lambda x: len(str(x)))
df_cleaned['body_word_count'] = df_cleaned['User Review Body_tokenized'].apply(lambda x: len(x) if isinstance(x, list) else 0)

print("New features 'title_char_count', 'title_word_count', 'body_char_count', and 'body_word_count' have been created.")
display(df_cleaned[['User Review Title_cleaned', 'title_char_count', 'title_word_count', 'User Review Body_cleaned', 'body_char_count', 'body_word_count']].head())
```

```python
import numpy as np
from scipy.sparse import hstack, csr_matrix


new_text_features = df_cleaned[['title_char_count', 'title_word_count', 'body_char_count', 'body_word_count']]
new_text_features_resampled, _ = smote.fit_resample(new_text_features, y)


numerical_features_resampled_sparse = csr_matrix(numerical_features_resampled)
new_text_features_resampled_sparse = csr_matrix(new_text_features_resampled)

X_combined = hstack([X_resampled, numerical_features_resampled_sparse, new_text_features_resampled_sparse])
print(f"Shape of X_combined: {X_combined.shape}")
```
